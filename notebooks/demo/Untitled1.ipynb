```python
from combinatorial_gwas.phenotypes.Field import Field
from combinatorial_gwas.data_catalog import get_catalog
```

```python
catalog = get_catalog()
```

```python
heritability_df = catalog.load("heritability_trait_level_summary")
heritability_df.head(3)
```

```python
display_cols = ['description', 'h2_liability', 'h2_sig', 'confidence', 'n_cases', 'n_controls', 'prevalence']
high_herit_traits = heritability_df.pheno.query(h2_sig = "in ['z7', 'z4']", confidence= "in ['medium', 'high']").sort_values("h2_liability", ascending = False)
high_herit_traits[display_cols].head(20)
```

```python
display_cols = ['description', 'h2_liability', 'h2_sig', 'confidence', 'n_cases', 'n_controls', 'prevalence']
high_herit_traits = heritability_df.pheno.query(h2_sig = "in ['z7', 'z4']", confidence= "in ['medium', 'high']", n_cases= ".notnull()").sort_values("h2_liability", ascending = False)
high_herit_traits[display_cols].head(20)
```
